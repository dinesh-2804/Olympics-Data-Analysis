## Handling Missing values

```python
df.isnull().sum()
```

### Observation

    - **Weight have 17101 missing values**

```python
kde=df.loc[:,'Age':'Weight']
```

```python
m=df.loc[:,'Age':]
for j in m.columns:
    plt.subplot(2,2,1)
    sns.kdeplot(m[j])
    plt.show()
```

### Observation


- The **Height** column shows a **roughly normal distribution**, so missing values were filled with the mean.
- The **Weight** and **Age columns** are **right-skewed**, meaning most participants are in the lower range of these values.
- This approach ensures that the data remains statistically consistent and less biased after imputation.


```python


df['Age'].fillna(df['Age'].median(), inplace=True)
```

### Obserbation 

-- Now we are filling the Age with the Median while checking the Age It is Right-Skewed

```python
df['Height'].fillna(df['Height'].mean(),inplace=True)
```

### Observation
- The missing values in the Height column were filled with the mean value because the distribution is symmetric (not skewed).
- Using the mean helps preserve the overall balance of the Height data.

```python
df['Weight'].fillna(df['Weight'].median(),inplace=True)
```

### Observation
- The missing values in the Weight column were filled using the median value, as the distribution is right-skewed.
- This approach maintains the data consistency and prevents bias due to extreme weight values.

```python
df['Medal'].fillna('No Medal',inplace=True)
```

### Observation
- The missing values in the Medal column were replaced with â€œNo Medalâ€ to represent non-medalist athletes.
- This provides a clear distinction between medal winners and participants without medals, improving the accuracy of categorical analysis.

# Visualizations

```python
sns.countplot(data=df, x='Medal', order=['Gold','Silver','Bronze','No Medal'], palette='pastel')
plt.title('Medal Distribution Among Athletes')
plt.xlabel('Medal')
plt.ylabel('Count')
plt.show()
```

### Observation

- The count plot shows the distribution of medals among all participating athletes.
- A large majority of athletes fall under the â€œNo Medalâ€ category, indicating that winning a medal is relatively rare in the Olympics.
- Among the medalists, Gold, Silver, and Bronze counts are fairly balanced, with slight variations depending on events and participation
  levels.

```python
sns.countplot(data=df,x='Season',hue='Medal')
plt.title('Season-Wise Medal Distribution')
plt.xlabel('Season')
plt.ylabel('Count')
plt.show()
```

### Observation

- The count plot illustrates the distribution of medals across the Summer and Winter Olympic seasons.
- It is evident that the Summer Olympics has a significantly higher number of participants and medals compared to the Winter Olympics.
- This is because the Summer Games include more sports and events, attracting a larger pool of athletes from around the world.
- The Winter Olympics, while smaller in scale, still show consistent medal achievements across its limited sports categories.

**Bivariate Analysis**
- Bivariate analysis examines the relationship between two variables â€” one may be numerical and the other categorical, or both may be
  numerical.
- It helps identify patterns, correlations, and dependencies that can lead to meaningful insights.

```python
top_sports = df['Sport'].value_counts().head(8).index
sns.countplot(data=df[df['Sport'].isin(top_sports)], x='Sport', hue='Sex')
plt.title('Gender Participation in Top Sports')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

### Observation:
- Male participation dominates in most sports, though female representation is increasing in events like Athletics and Swimming.
- This reveals gender balance trends across different sports categories.

```python
sns.boxplot(data=df[df['Medal'] != 'No Medal'], x='Medal', y='Weight', palette='pastel')
plt.title('Weight Distribution Across Medal Types')
plt.xlabel('Medal Type')
plt.ylabel('Weight (kg)')
plt.show()
```

### Observation:
- The median weights of medalists donâ€™t differ drastically, but variability is seen depending on the sport type.
- For example, heavier athletes are often seen in power-based sports (weightlifting), while lighter ones in endurance events.
- No strong correlation between weight and medal type, but the spread reflects sport diversity.

```python
sns.countplot(data=df, x='Sex', hue='Medal', palette='cool')
plt.title('Gender-wise Medal Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## Multivariate Analysis
- Multivariate analysis studies the combined effect of multiple variables â€” numerical and categorical â€” to understand patterns that arenâ€™t visible from pairwise (bivariate) comparisons.

```python
plt.figure(figsize=(10,4))
sns.scatterplot(data=df,x='Height',y='Weight',hue='Sex')
plt.title('Relationship between Height and Weight of Athletes by Gender')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```
